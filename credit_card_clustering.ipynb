```python
mean_value=df['CREDIT_LIMIT'].mean()
df['CREDIT_LIMIT']=df['CREDIT_LIMIT'].fillna(mean_value)
```

```python
mean_value=df['MINIMUM_PAYMENTS'].mean()
df['MINIMUM_PAYMENTS']=df['MINIMUM_PAYMENTS'].fillna(mean_value)
```

```python
df.corr()
```

```python
df = df.drop('CUST_ID', axis=1)
df.head(3)
```

```python
#apply scaling means standardlisation  means apply StandardScaler (Z-score formula)
from sklearn.preprocessing import StandardScaler
#Apply standardization means same the scaling/unit  of both column Age and Income column
#create the object of StandardScaler class
ss=StandardScaler()

##input always 2D array means 2 Dimensional array [[]]
for col in df:
    df[col]=ss.fit_transform(df[[col]])
```

```python
df.head()
```

```python
#call class KMeans inbuilt call
from sklearn.cluster import KMeans  #Kmeans inbuilt class
```

```python
#Elbow method use :  How many clusters in our dataset (mall.csv)
k_rng=range(1,21) #Start =1 stop=21-1=20 step-=+1
#declare empty list for sum of squared error SSE
SSE=[]#declare empty SSE list object
k_cluster=[]#declare empty No. of cluster list object
for k in k_rng:
    #Create the object of KMeans class
    km=KMeans(n_clusters=k)  #first time k=1 then second time k=2
    #We train the model use fit()
    km.fit(df)
    ##inertia_ inbuilt method of Kmeans class which calculate the sum of squared error
    #formula of SSE=sum(centerpoint-datapoint)^2
    SSE.append(km.inertia_)
    k_cluster.append(k)
    
print("No. of cluster K  : Sum of squared error : ")
print( k_cluster,' \n ',SSE)
```

```python
#Visualize SSE and k
plt.figure(figsize=(8,8))
plt.scatter(k_rng,SSE,color="red")
plt.plot(k_rng,SSE) #line plot
plt.xticks(k_rng)
plt.xlabel("Value Of K")
plt.ylabel("Sum of squared Error SSE")
plt.show()
```

```python
predicted=km.fit_predict(df) #predicted new output variable
print(predicted)
```

```python
#add new column after fit_predict in dataset df
#df["Newcolumn"]=newvalue
df["Cluster"]=predicted  #here Cluster new column which add existing data frame df and hold the 
#value of predicted variable
df.head(10)
```
