```python
import requests
from bs4 import BeautifulSoup
import csv
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import plotly.express as px

```

```python
soup=BeautifulSoup(requests.get("https://www.worldometers.info/world-population/india-population/").text,'lxml')
```

```python
soup
```

```python
table=soup.find("table",class_="table table-striped table-bordered table-hover table-condensed table-list")
```

```python
heading=[h.text.replace("\n","").replace("\xa0","")for h in table.find_all('th')]
```

```python
heading
```

```python
rows=[r for r in table.find_all('tr') if not r.has_attr('data-continent')]
```

```python
rows
```

```python
rows=rows[1:]
```

```python
result=[]
```

```python
result.append(heading)
```

```python
for table_rows in rows:
    row=[]
    for data in table_rows.find_all('td'):
        if data.a is not None: row.append(data.a.text)
        else: row.append(data.text)
    result.append(row)
```

```python
csv_file = open("world_population","w",newline="")
```

```python
writer=csv.writer(csv_file)
```

```python
writer.writerows(result)
```

```python
csv_file.close()
```

```python
world=world.rename(columns={"Country's Share of World Pop":"Country_share_of_world_pop"})
```

```python
cols=['Year', 'Population', 'Yearly %  Change', 'Yearly Change',
       'Migrants (net)', 'Median Age', 'Fertility Rate', 'Density (P/Km²)',
       'Urban Pop %', 'Urban Population', 'Country_share_of_world_pop',
       'World Population', 'IndiaGlobal Rank']
    

world[cols] = world[cols].replace({ ',': '','%':""}, regex=True)
```

```python
for label in world.columns:
    if label != 'Year' :
        world[label]=pd.to_numeric(world[label],errors='coerce').convert_dtypes() 
```
